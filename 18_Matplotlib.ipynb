# Matplotlib.Pyplot

**matplotlib.pyplot** is a collection of command style functions that make matplotlib work like MATLAB

Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a fugure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel('Numbers')
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

If you provide a single list or array to the plot() command, matplotlib assused ut is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0. Hence the x data are [0,1,2,3]

### plot x vs y

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel('Squares')
plt.xlabel('Numbers')
plt.grid() #grid on
plt.show()
```

For every x, y pair of arguments, there is an optional third argument which is the **format string** that inicates the color and line type of the plot.

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.ylabel('Squares')
plt.xlabel('Numbers')
plt.grid() #grid on
plt.show()
```

If matplotlib were limited to working with lists, it would be fairly useless for numeric processing. Generally, you will use **numpy arrays**. In fact, all sequences are converted to numpy arrays internally.

```python
import numpy as np
```

```python
t = np.arange(0.,5.,0.2)

#blue dashes, red squares and green triangles
plt.plot(t, t**2, 'b--', label= '^2') # 'rs', 'g^'
plt.plot(t, t**2.2, 'rs', label= '^2.2')
plt.plot(t, t**2.5, 'g^', label= '^2.5')
plt.grid()
plt.legend() #add legend based on the line labels
plt.show()
```

# Controlling line properties

#### use keyword args

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x, y, linewidth=5.0 )
plt.show()
```

#### use the setp()

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

lines = plt.plot(x1, y1, x2, y2)

#use keyword args
plt.setp(lines[0],color='r',linewidth=2.0)

#or MATLAB style string value pairs
plt.setp(lines[1],'color','g','linewidth',2.0)

plt.grid()
plt.show()
```

# Working with multiple figures and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
#The subplot() command specifies numrows, numcols,
#fignum where fignum ranges from 1 to numrows * numcols.

plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b-')

plt.subplot(212)
plt.grid()
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
```

```python
plt.figure(1) # the first figure

plt.subplot(211) # the first subplot in first figure
plt.plot([1,2,3])
plt.subplot(212) # the second subplot in first figure
plt.plot([4,5,6])

plt.figure(2) # a second figure
plt.plot([4,5,6]) # creates a subplot(111) by default

plt.figure(1) # figure 1 current: subplt(212) still current
plt.subplot(211) # make subplot(211) in figure1 current
plt.title('easy as 1,2,3') # subplot 211 title
plt.show()
```
